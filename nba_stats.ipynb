```python
# Takım istatistiklerini topla
team_stats = players.groupby('Tm')[['PTS', 'AST', 'TRB']].sum().reset_index()

# Takımları toplam sayıya göre sırala
team_stats_sorted_pts = team_stats.sort_values(by='PTS', ascending=False)

# Takımları toplam asiste göre sırala
team_stats_sorted_ast = team_stats.sort_values(by='AST', ascending=False)

# Takımları toplam ribaunda göre sırala
team_stats_sorted_trb = team_stats.sort_values(by='TRB', ascending=False)

print("Toplam Sayıya Göre Sıralı Takım İstatistikleri:")
print(team_stats_sorted_pts)

print("\nToplam Asiste Göre Sıralı Takım İstatistikleri:")
print(team_stats_sorted_ast)

print("\nToplam Ribaunda Göre Sıralı Takım İstatistikleri:")
print(team_stats_sorted_trb)
```

```python
# Sıralı takım istatistiklerini görselleştir
plt.figure(figsize=(13, 6))

plt.subplot(1, 3, 1)
sns.barplot(x='PTS', y='Tm', data=team_stats_sorted_pts)
plt.title('Toplam Sayıya Göre Sıralı Takımlar')

plt.subplot(1, 3, 2)
sns.barplot(x='AST', y='Tm', data=team_stats_sorted_ast)
plt.title('Toplam Asiste Göre Sıralı Takımlar')

plt.subplot(1, 3, 3)
sns.barplot(x='TRB', y='Tm', data=team_stats_sorted_trb)
plt.title('Toplam Ribaunda Göre Sıralı Takımlar')

plt.tight_layout()
plt.show()
```
